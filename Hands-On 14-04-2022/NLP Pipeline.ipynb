### Making Necessary Imports

```python
import spacy
import textacy
```

### Loading NLP Model

```python
# Load the large English NLP model
nlp = spacy.load('en_core_web_lg')
```

### Reading File

```python
# Read file
f = open('test_file.txt','r')
text = f.read()
f.close()
```

```python
tokens = nlp(text)
tokens
```

### Sentence Segmentation

```python
# Sentence Segmentation
n_sent = 0
for sentence in tokens.sents:
    n_sent = n_sent + 1
    print(sentence.text)

# number of sentences
print('='*50)
print("Number of sentences: ",n_sent)
```

###  Tokenization

```python
# Word tokenization
for token in tokens:
    print('[',token.text,']',end = "")
    
# Number of tokens
print()
print("="*50)
print("Number of tokens = ",len(tokens))
```

### Parts of Speech Prediction

```python
# Identification of each token as some part of speech
# 1. Noun
# 2. Pronoun
# 3. Adjective
# 4. Verb
# 5. Adverb
# 6. Prepositions
# 7. Conjunctions
# 8. Articles
```

```python
for token in tokens:
    print("-"*50)
    print("TOKEN: ",token.text)
    print("PART OF SPEECH: ",token.pos_)
    print("EXPLAINATION: ",spacy.explain(token.pos_))
```

### Lemmatization

```python
for token in tokens:
    print("-"*50)
    print("TOKEN: ",token.text)
    print("LEMMA: ",token.lemma_)
```

###  Stop Words

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
# creating list of stopwords
stopwords = spacy.lang.en.STOP_WORDS
print(stopwords)
```

```python
# removing stopwords
mytokens = []
mywords = []
for token in tokens:
    if (token.is_stop == False):
        mytokens.append(token)
        mywords.append(token.text)
```

```python
# After removal of stopwords
print((" ".join(mywords)))
```

### Step 6: Dependency Parsing

```python
from spacy import displacy
```

```python
options = {"compact": True, "bg": "#09a3d5",
           "color": "white", "font": "Source Sans Pro"}
displacy.render(tokens, style = "dep", options = options, jupyter = True)
```

###  Noun Phrases

```python
# noun phrases/chunks
noun_phrases = textacy.extract.noun_chunks(tokens) 
```

```python
noun_phrases = map(str, noun_phrases)
noun_phrases = map(str.lower, noun_phrases)
```

```python
all_phrases = []
for phrase in noun_phrases:
  # length of noun phrase
  if len(phrase.split(' ')) > 1:
    all_phrases.append(phrase)
```

```python
# removinf duplicates
all_phrases = set(all_phrases)
print(all_phrases)
```
