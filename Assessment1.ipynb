## Alcohol Consumption 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/student-mat.csv'
df = pd.read_csv(url)
df.head()
```

```python
alc = df.loc[:,'school':'guardian']
alc.head()
```

```python
caps = lambda l : l.capitalize() 
```

```python
alc['Mjob'] = alc['Mjob'].apply(caps)
alc['Fjob'] = alc['Fjob'].apply(caps)
```

```python
alc.head()
```

```python
def majority(age):
  if age > 17:
    return True
  else:
    return False
```

```python
alc['legal_drinker'] = alc['age'].apply(majority)
```
